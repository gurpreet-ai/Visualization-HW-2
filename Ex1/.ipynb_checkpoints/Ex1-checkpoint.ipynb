###     1) Carapace Width [mm] V. Carapace Length [mm]

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib as mpl

data = np.recfromcsv('crabs.csv', usecols=(1, 2, 3, 4, 5, 6, 7, 8))

CL = [i for i in data['cl']]
CW = [i for i in data['cw']]

gs   = mpl.gridspec.GridSpec(1, 1)
fig = plt.figure(figsize=(12, 8))

ax   = fig.add_subplot(gs[0])

ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_linewidth(0.5)
ax.xaxis.set_ticks_position('bottom')

ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(0.5)
ax.yaxis.set_ticks_position('left')

ax.xaxis.set_tick_params('major', length=2, labelsize=10)
ax.yaxis.set_tick_params('major', length=2, labelsize=10)

ax.set_xlim((10,50))
ax.set_ylim((10,60))

ax.set_xlabel(r'$Carapace  Length [mm]$', fontsize=12, family='sans-serif')
ax.set_ylabel(r'$Carapace  Width [mm]$', fontsize=12, family='sans-serif')
ax.set_title(r'$Carapace  Width [mm]$  V.  $Carapace  Length [mm]$', fontsize=15, family='sans-serif')

ax.plot(CL, CW, marker='.', color='k', linestyle='',
    markersize=4, clip_on=False)

ax.autoscale_view()

fig.savefig('figure_1.png')
```

###     2) Carapace length [mm] and Frontal Lobe Size [mm] V. Carapace Width [mm]

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib as mpl

data = np.recfromcsv('crabs.csv', usecols=(1, 2, 3, 4, 5, 6, 7, 8))

CL = [i for i in data['cl']]
CW = [i for i in data['cw']]
FL = [i for i in data['fl']]

gs   = mpl.gridspec.GridSpec(1, 2)
fig = plt.figure(figsize=(16, 10))

ax = fig.add_subplot(gs[0])

ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')

ax.xaxis.set_tick_params('major', length=2, labelsize=10)
ax.yaxis.set_tick_params('major', length=2, labelsize=10)

ax.set_xlim((10,50))
ax.set_ylim((10,60))

ax.set_xlabel('Carapace  Width [mm]', fontsize=12, family='sans-serif')
ax.set_ylabel('Carapace  Length [mm]', fontsize=12, family='sans-serif')
ax.set_title('CL and FL V. CW', fontsize=15, family='sans-serif')

a = ax.plot(CW, CL, marker='.', color='blue', linestyle='',
        markersize=4, clip_on=False)

ax.legend(a,['CL V CW'], loc="upper left")

ax2 = ax.twinx()
ax2.set_ylabel('Frontal  Lobe Size [mm]', fontsize=12, family='sans-serif')
ax2.yaxis.set_tick_params('major', length=2, labelsize=10)
ax2.set_ylim(0, 60)
ax2.set_xlim(10, 60)

b = ax2.plot(CW, FL, marker='.', color='green', linestyle='',
    markersize=4, clip_on=False)

ax2.legend(b, ['FL V CW'], loc="upper right")

ax.autoscale_view()
ax2.autoscale_view()

bx = fig.add_subplot(gs[1])

bx.xaxis.set_tick_params('major', length=2, labelsize=10)
bx.yaxis.set_tick_params('major', length=2, labelsize=10)

bx.set_xlim((10,60))
bx.set_ylim((0,30))
bx.autoscale_view()

bx.set_xlabel('CL carapace Width [mm]', fontsize=12, family='sans-serif')
bx.set_ylabel('Frontal  Lobe  Size [mm]', fontsize=12, family='sans-serif')
bx.set_title('CL and FL V. CW', fontsize=15, family='sans-serif')

bx.scatter(CW, FL, s=CL, alpha=1)
bx.autoscale_view()

fig.savefig('figure_2.svg')
```

The above graphs shows Carapace Width V. Carapace Length and Carapace Width V. Frontal Lobe Size. The first graph uses double y-axes to show the data dimensions. The second graph shows represent the third variable (CL) with the circles of different sizes. The third dimension in the second graph is hard to read because you can't tell the difference in the sizes of the circles. 

The third dimension using color scale is shown below.

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib as mpl

data = np.recfromcsv('crabs.csv', usecols=(1, 2, 3, 4, 5, 6, 7, 8))

CL = [i for i in data['cl']]
CW = [i for i in data['cw']]
FL = [i for i in data['fl']]

gs   = mpl.gridspec.GridSpec(1, 1)
fig = plt.figure(figsize=(12, 8))

bx = fig.add_subplot(gs[0])

bx.xaxis.set_tick_params('major', length=2, labelsize=10)
bx.yaxis.set_tick_params('major', length=2, labelsize=10)

bx.set_xlim((10,60))
bx.set_ylim((0,30))
bx.autoscale_view()

bx.set_xlabel('CL carapace Width [mm]', fontsize=12, family='sans-serif')
bx.set_ylabel('Frontal  Lobe  Size [mm]', fontsize=12, family='sans-serif')
bx.set_title('CL and FL V. CW', fontsize=15, family='sans-serif')

cmap = plt.get_cmap('hot')
a = bx.scatter(CW, FL, c=CL, cmap=cmap)
cax = fig.colorbar(a)
cax.set_label("CL carapace Length [mm]")

bx.autoscale_view()

fig.savefig('figure_3.png')
```

The graph above uses color to create the third dimension and it makes it very easy to compare the measurements shown. I think this graph is the most effective out of the three and it's accurate. The user can get close the exact value of the CL using the color bar.

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib as mpl

data = np.recfromcsv('crabs.csv', usecols=(1, 2, 3, 4, 5, 6, 7, 8))

CL = [i for i in data['cl']]
CW = [i for i in data['cw']]
FL = [i for i in data['fl']]

gs   = mpl.gridspec.GridSpec(1, 1)
fig = plt.figure(figsize=(14, 10))

bx = fig.add_subplot(gs[0])

bx.spines['top'].set_visible(False)
bx.spines['bottom'].set_linewidth(0.5)
bx.xaxis.set_ticks_position('bottom')

bx.spines['right'].set_visible(False)
bx.spines['left'].set_linewidth(0.5)
bx.yaxis.set_ticks_position('left')

bx.xaxis.set_tick_params('major', length=2, labelsize=10)
bx.yaxis.set_tick_params('major', length=2, labelsize=10)

#bx.set_xlim((10,60))
#bx.set_ylim((0,30))
bx.autoscale_view()

bx.set_xlabel('CW carapace Width [mm]', fontsize=12, family='sans-serif')
bx.set_ylabel('Frontal  Lobe  Size [mm]', fontsize=12, family='sans-serif')
bx.set_title('CL and FL V. CW (with sorting CW)', fontsize=15, family='sans-serif')

CL_Blue_male   = [i['cl'] for i in data if i['sp'] == 'B' and i['sex'] == 'M']
CL_Blue_female = [i['cl'] for i in data if i['sp'] == 'B' and i['sex'] == 'F']
CL_Oran_male   = [i['cl'] for i in data if i['sp'] == 'O' and i['sex'] == 'M']
CL_Oran_female = [i['cl'] for i in data if i['sp'] == 'O' and i['sex'] == 'F']

CW_Blue_male   = [i['cw'] for i in data if i['sp'] == 'B' and i['sex'] == 'M']
CW_Blue_female = [i['cw'] for i in data if i['sp'] == 'B' and i['sex'] == 'F']
CW_Oran_male   = [i['cw'] for i in data if i['sp'] == 'O' and i['sex'] == 'M']
CW_Oran_female = [i['cw'] for i in data if i['sp'] == 'O' and i['sex'] == 'F']

FL_Blue_male   = [i['fl'] for i in data if i['sp'] == 'B' and i['sex'] == 'M']
FL_Blue_female = [i['fl'] for i in data if i['sp'] == 'B' and i['sex'] == 'F']
FL_Oran_male   = [i['fl'] for i in data if i['sp'] == 'O' and i['sex'] == 'M']
FL_Oran_female = [i['fl'] for i in data if i['sp'] == 'O' and i['sex'] == 'F']

CW_Blue_male, CL_Blue_male, FL_Blue_male  = zip(*sorted(zip(      CW_Blue_male,   CL_Blue_male,   FL_Blue_male)))
CW_Blue_female, CL_Blue_female, FL_Blue_female  = zip(*sorted(zip(CW_Blue_female, CL_Blue_female, FL_Blue_female)))
CW_Oran_male, CL_Oran_male, FL_Oran_male  = zip(*sorted(zip(      CW_Oran_male,   CL_Oran_male,   FL_Oran_male)))
CW_Oran_female, CL_Oran_female, FL_Oran_female  = zip(*sorted(zip(CW_Oran_female, CL_Oran_female, FL_Oran_female)))

bx.scatter(CW_Blue_male, FL_Blue_male, s=CL_Blue_male, alpha=.2, color="Navy", label="Blue Male")
bx.scatter(CW_Blue_female, FL_Blue_female, s=CL_Blue_female, alpha=.2, color="Aqua", label="Blue female")
bx.scatter(CW_Oran_male, FL_Oran_male, s=CL_Oran_male, alpha=.2, color="Orange", label="Orange Male")
bx.scatter(CW_Oran_female, FL_Oran_female, s=CL_Oran_female, alpha=.2, color="OrangeRed", label="Orange female")

bx.plot(CW_Blue_male, FL_Blue_male)
bx.plot(CW_Blue_female, FL_Blue_female)
bx.plot(CW_Oran_male, FL_Oran_male)
bx.plot(CW_Oran_female, FL_Oran_female)

bx.legend(loc='upper left')
bx.autoscale_view()

fig.savefig('figure_5_A.png')
```

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib as mpl

data = np.recfromcsv('crabs.csv', usecols=(1, 2, 3, 4, 5, 6, 7, 8))

CL = [i for i in data['cl']]
CW = [i for i in data['cw']]
FL = [i for i in data['fl']]

gs   = mpl.gridspec.GridSpec(1, 1)
fig = plt.figure(figsize=(12, 8))

ax = fig.add_subplot(gs[0])

ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')

ax.xaxis.set_tick_params('major', length=2, labelsize=10)
ax.yaxis.set_tick_params('major', length=2, labelsize=10)

ax.set_xlim((5,22))
ax.set_ylim((0,50))

BD_Blue_male   = [i['bd'] for i in data if i['sp'] == 'B' and i['sex'] == 'M']
BD_Blue_female = [i['bd'] for i in data if i['sp'] == 'B' and i['sex'] == 'F']
BD_Oran_male   = [i['bd'] for i in data if i['sp'] == 'O' and i['sex'] == 'M']
BD_Oran_female = [i['bd'] for i in data if i['sp'] == 'O' and i['sex'] == 'F']

CL_Blue_male   = [i['cl'] for i in data if i['sp'] == 'B' and i['sex'] == 'M']
CL_Blue_female = [i['cl'] for i in data if i['sp'] == 'B' and i['sex'] == 'F']
CL_Oran_male   = [i['cl'] for i in data if i['sp'] == 'O' and i['sex'] == 'M']
CL_Oran_female = [i['cl'] for i in data if i['sp'] == 'O' and i['sex'] == 'F']

RW_Blue_male   = [i['rw'] for i in data if i['sp'] == 'B' and i['sex'] == 'M']
RW_Blue_female = [i['rw'] for i in data if i['sp'] == 'B' and i['sex'] == 'F']
RW_Oran_male   = [i['rw'] for i in data if i['sp'] == 'O' and i['sex'] == 'M']
RW_Oran_female = [i['rw'] for i in data if i['sp'] == 'O' and i['sex'] == 'F']

ax.set_xlabel('RW Rear Width [mm]', fontsize=12, family='sans-serif')
ax.set_ylabel('BD Body Depth [mm]', fontsize=12, family='sans-serif')
ax.set_title('BD and CL V. RW', fontsize=15, family='sans-serif')

a, = ax.plot(RW_Blue_male, CL_Blue_male, marker='.', color='blue', linestyle='',
        markersize=10, clip_on=False, label="CL_B_male V. RW_Blue_male")

b, = ax.plot(RW_Blue_female, CL_Blue_female, marker='.', color='dodgerblue', linestyle='',
        markersize=10, clip_on=False, label="CL_B_female V. RW_Blue_female")

c, = ax.plot(RW_Oran_male, CL_Oran_male, marker='.', color='orange', linestyle='',
        markersize=10, clip_on=False, label="CL_O_male V. RW_Oran_male")

d, = ax.plot(RW_Oran_female, CL_Oran_female, marker='.', color='orangered', linestyle='',
        markersize=10, clip_on=False, label="CL_O_female V. RW_Oran_female")


ax.legend([a, b, c, d], loc="upper left")

ax2 = ax.twinx()
ax2.set_ylabel('CL carapace Width [mm]', fontsize=12, family='sans-serif')
ax2.yaxis.set_tick_params('major', length=2, labelsize=10)
ax2.set_ylim(0, 50)
ax2.set_xlim(5, 25)

e, = ax2.plot(RW_Blue_male, BD_Blue_male, marker='.', color='mediumblue', linestyle='',
        markersize=10, clip_on=False, label="BD_B_male V. RW_Blue_male")

f, = ax2.plot(RW_Blue_female, BD_Blue_female, marker='.', color='royalblue', linestyle='',
        markersize=10, clip_on=False, label="BD_B_female V. RW_Blue_female")

g, = ax2.plot(RW_Oran_male, BD_Oran_male, marker='.', color='darkred', linestyle='',
        markersize=10, clip_on=False, label="BD_O_male V. CL_Oran_male")

h, = ax2.plot(RW_Oran_female, BD_Oran_female, marker='.', color='lightsalmon', linestyle='',
        markersize=10, clip_on=False, label="BD_O_female V. CL_Oran_female")

ax2.legend([e, f, g, h], loc="lower right")

ax.autoscale_view()
ax2.autoscale_view()


fig.savefig('figure_8.png')
```

The graph above is easy to read because of the 3rd varible on the 2nd y-axis.
